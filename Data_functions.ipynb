```python
data.head()
```

* **df.tail()** : By using this function we get the last 5 rows of data set.

```python
data.describe()
```

```python
data['t1'].mean()
```

* df[' '].var() : to find the variance of paricular column.

```python
data['t1'].var()
```

```python
data.loc[0]
```

* **data.iloc[ ]**: to find the both row and column value by giving the index value.

```python
import pandas as pd
data = pd.read_csv("/content/titanic-dataset/SVMtrain.csv")
data.head()
```

```python
# prompt: generate a code for missing values and duplicates in the dataset

import pandas as pd

# Assuming 'data' DataFrame is already loaded as in the previous code

# Check for missing values
missing_values = data.isnull().sum()
print("Missing values:\n", missing_values)

# Check for duplicate rows
duplicate_rows = data.duplicated().sum()
print("\nNumber of duplicate rows:", duplicate_rows)

# Handle missing values (example: fill with mean for numerical, mode for categorical)
data['Age'].fillna(data['Age'].mean(), inplace=True)
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)

# Handle duplicates (example: remove duplicate rows)
data.drop_duplicates(inplace=True)

# Verify the changes
missing_values_after = data.isnull().sum()
print("\nMissing values after handling:\n", missing_values_after)

duplicate_rows_after = data.duplicated().sum()
print("\nNumber of duplicate rows after handling:", duplicate_rows_after)
```

```python
# prompt: generate a code for to give dropdown for Parch

import ipywidgets as widgets
from IPython.display import display

# Assuming 'data' DataFrame is already loaded and preprocessed

# Get unique values from the 'Parch' column
unique_parch_values = sorted(data['Parch'].unique())

# Create a dropdown widget
parch_dropdown = widgets.Dropdown(
    options=unique_parch_values,
    value=unique_parch_values[0],  # Set the default value
    description='Parch:'
)

# Display the dropdown
display(parch_dropdown)


# Example of how to use the selected value
def on_parch_change(change):
    selected_parch = change['new']
    print(f"Selected Parch: {selected_parch}")
    # Add your code here to filter the data based on the selected 'Parch' value
    # For example:
    filtered_data = data[data['Parch'] == selected_parch]
    print(filtered_data)

parch_dropdown.observe(on_parch_change, names='value')
```
